```python
payload = {
    "type": "tree",
}

tree = await websocket_test(payload, indent=2)
```

```python
for obj in tree:
    if obj["objectInfo"]["id"] != 0:
        payload = {
            "type": "hierarchy",
            "command": "attach",
            "childId": obj["objectInfo"]["id"],
            "parentId": 0,
        }
        await websocket_test(payload)
```
